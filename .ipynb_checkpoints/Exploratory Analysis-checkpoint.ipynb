```python
%pylab inline
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/uber-tlc-foil-response/master/uber-trip-data/uber-raw-data-apr14.csv")
```

```python
data.head()
```

```python
#data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
date_time_split = data['Date/Time'].str.split(' ')
```

```python
date_time_split.str.get(0)
```

```python
data['Date'] = date_time_split.str.get(0)
data['Time'] = date_time_split.str.get(1)

```

```python
data.head()
```

```python
data.shape
```

```python
dt = '4/1/2014 0:17:00'
m, d, y = dt.split('/')
```

```python
dt = pd.to_datetime(dt)
```

```python
dt.day_name()
```

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
data.describe()
```

```python
data.head()
```

```python
data['dom'] = data['Date/Time'].map(lambda s: s.day)
```

```python
data.head()
```

```python
data['weekday'] = data['Date/Time'].map(lambda s: s.weekday())
data['hour'] = data['Date/Time'].map(lambda s: s.hour)
```

```python
data.sample(10)
```

```python
data = data.drop(columns = ['Date','Time'])
```

```python
data.head()
```

```python
hist(data.dom, bins=30,width=0.5, range=(-.5,30.5),weights=None)
xlabel('Date of the month')
ylabel('Frequency')
title('Frequency by DOM')
```

```python
data.shape
```

```python
#for k, rows in data.groupby('dom'):
 #   print((k, len(rows)))
    
by_date = data.groupby('dom').apply(lambda s: len(s))
```

```python
by_date
```

```python
plot(by_date)
```

```python
bar(range(1,31),by_date)
```

```python
by_date_sorted = by_date.sort_values()
```

```python
bar(range(1,31),by_date_sorted)
xticks(range(1,31), by_date_sorted.index);
xlabel('Date of the month')
ylabel('Frequency')
title('Frequency by DOM')
```
